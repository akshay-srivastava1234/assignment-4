```python

geometry = [Point(xy) for xy in zip(incident_df["lng"], incident_df["lat"])]
```

```python
geo_df = gpd.GeoDataFrame(incident_df,
                            crs = 4326,
                            geometry = geometry)

geo_df.head()
```

```python
geo_df.shape
```

```python
geo_df_map = geo_df.plot(figsize=(15,15), markersize = 5)
geo_df_map.set_axis_off()
```

```python
geo_df.shape()
```

```python
geo_df.crs
```

```python
fig, ax = plt.subplots(figsize =(15,15))

geo_df[geo_df['text_general_code'] == 'Thefts'].plot(ax = ax, markersize = 20, color = "blue", marker = "o", label = "Thefts")
geo_df[geo_df['text_general_code'] == 'Robbery No Firearm'].plot(ax = ax, markersize = 20, color = "red", marker = "^", label = "Robbery")
geo_df[geo_df['text_general_code'] == 'Burglary Residential'].plot(ax = ax, markersize = 20, color = "orange", marker = "^", label = "Burglary Residential")
geo_df[geo_df['text_general_code'] == 'Burglary Non-Residential'].plot(ax = ax, markersize = 20, color = "orange", marker = "o", label = "Burglary Non-Residential")
geo_df[geo_df['text_general_code'] == 'Robbery Firearm'].plot(ax = ax, markersize = 20, color = "red", marker = "o", label = "Robbery Firearm")
geo_df[geo_df['text_general_code'] == 'Aggravated Assault No Firear'].plot(ax = ax, markersize = 20, color = "purple", marker = "o", label = "Aggravated Assault No Firear")
geo_df[geo_df['text_general_code'] == 'Aggravated Assault Firearm'].plot(ax = ax, markersize = 20, color = "purple", marker = "^", label = "Aggravated Assault Firearm")
geo_df[geo_df['text_general_code'] == 'Theft from Vehicle'].plot(ax = ax, markersize = 20, color = "pink", marker = "o", label = "Theft from Vehicle")
geo_df[geo_df['text_general_code'] == 'Rape'].plot(ax = ax, markersize = 20, color = "pink", marker = "^", label = "Rape")
geo_df[geo_df['text_general_code'] == 'Vandalism/Criminal Mischief'].plot(ax = ax, markersize = 20, color = "yellow", marker = "^", label = "Vandalism/Criminal Mischief")
geo_df[geo_df['text_general_code'] == 'All Other Offenses'].plot(ax = ax, markersize = 20, color = "yellow", marker = "o", label = "All Other Offenses")
geo_df[geo_df['text_general_code'] == 'Other Assaults'].plot(ax = ax, markersize = 20, color = "green", marker = "o", label = "Other Assaults")




#geo_df.plot(column = 'BUILD_RANK', ax = ax, alpha = 0.5, legend = True, markersize = 10)
plt.legend(prop = {'size' : 15})
#plt.show()
```
